# Comparaison de la production d'energie mondiale

```python
Gas_mtoe = df_gsprod[df_gsprod["Country"].str.contains("Total World")].rename(columns ={"Country":"Energy Type"})
Coal_mt = Coal_mtoe[Coal_mtoe["Country"].str.contains("Total World")].rename(columns ={"Country":"Energy Type"})
df_renewable_consumption_tWorld = df_renewable_consumption[df_renewable_consumption["Country"].str.contains("Total World")].rename(columns ={"Country":"Energy Type"})
df_nuclear = df_nuclear_consumption[df_nuclear_consumption["Country"].str.contains("Total World")].rename(columns ={"Country":"Energy Type"})
Oil_Mtonnes = oilmt[oilmt["Country"].str.contains("Total World")].rename(columns ={"Country":"Energy Type"})

Gas_mtoe.loc[71, 'Energy Type'] = 'Gas'
Oil_Mtonnes.loc[71,'Energy Type'] = 'Oil'
Coal_mt.loc[55,'Energy Type'] = 'Coal'
df_nuclear.iloc[0,0]='Nuclear'
df_renewable_consumption_tWorld.loc[107,'Energy Type'] = 'Renewable'

Mondiale = df_renewable_consumption_tWorld.append(Coal_mt).append(Oil_Mtonnes).append(Gas_mtoe).append(df_nuclear_consumption_world).append(df_nuclear)
Mondiale.set_index("Energy Type").T.plot(title="Evolution Production Energie Mondiale",figsize=(8,8))
```
